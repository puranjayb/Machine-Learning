```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("train.csv")
df1 = pd.read_csv("test.csv")
df = df.dropna()
df1 = df1.dropna()
```

```python
print(df.info)
print(df1.info)
```

```python
#Data for training the model
xTrain = np.array( df.x ).reshape(699,1)
yTrain = np.array( df.y ).reshape(699,1)

#Data for testing the model
xTest = np.array( df1.x ).reshape(300,1)
yTest = np.array( df1.y ).reshape(300,1)
```

```python
reg = LinearRegression()
reg.fit(xTrain, yTrain)
```

```python
prediction = reg.predict(xTest)
```

```python
square_error = mean_squared_error(yTest, prediction)
square_error
```

```python
plt.xlabel('Area')
plt.ylabel('Price')

plt.scatter(xTest, yTest, color='green', marker='+')
plt.scatter(xTest, prediction, color='red', marker='*')
```
